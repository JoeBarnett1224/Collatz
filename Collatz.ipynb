```python
import multiprocess
import tkinter as tk

def collatz(n):
    n = int(n)
    sequence = [n]
    while n >= 1:
        if n == 1:
            n = 0
        elif n % 2 == 1:
            sequence.append(int(3 * n + 1))
            n = 3 * n + 1
        else:
            sequence.append(int(n / 2))
            n = n / 2
    return sequence

def string_counter(list1):
    i = 0
    boolean = True
    while i < len(list1):
        if len(list1[i]) > 18 or int(list1[i]) <= 0:
            boolean = False
        i += 1
    return boolean

class App(tk.Tk):
    def __init__(self, title, state):
        
        #Create Window
        
        super().__init__()
        self.title(title)
        self.state(state)
        
        #Title

        self.title_label = tk.Label(self, text = "Collatz", font = "Calibri 24 bold")
        self.title_label.pack()
        
        #Input

        self.input_frame = tk.Frame(self)
        self.input_frame.pack()
        self.input_label = tk.Label(self.input_frame,
         text = "Positive integer(s) (separated by commas): ", font = "Calibri 16")
        self.input_label.pack(side = "left")
        self.input_entry_variable = tk.StringVar()
        self.input_entry = tk.Entry(self.input_frame, font = ("Calibri 16"),
         textvariable = self.input_entry_variable)
        self.input_entry.pack(side = "left")
        
        #Submit

        self.submit_button = tk.Button(self, text = "Submit", command = self.submit)
        self.submit_button.pack()

        #Output

        self.output_string = tk.StringVar()
        self.output_label = tk.Label(self, textvariable = self.output_string,
         font = "Calibri 16")
        self.output_label.pack()
        
        #signature

        self.signature_frame = tk.Frame(self)
        self.signature_frame.pack(side = "bottom")
        self.signature_label = tk.Label(self.signature_frame,
         text = "Created by Joseph W Barnett", font = "Calibri 10")
        self.signature_label.pack()
        
        #Help & Acknowledgements

        self.bottom_frame = tk.Frame(self)
        self.bottom_frame.pack(side = "bottom", fill = "x")
        self.help_button = tk.Button(self.bottom_frame, text = "Help", command = self.help1)
        self.help_button.pack(side = "left")
        self.acknowledgement_button = tk.Button(self.bottom_frame, text = "Acknowledgements",
         command = self.acknowledgement)
        self.acknowledgement_button.pack(side = "right")
        
        #Run
        
        self.mainloop()
        
    def submit(self):
        n_list = self.input_entry_variable.get()
        n_list = n_list.strip(" ")
        n_list = n_list.split(",")
        if len(n_list) > 20:
            self.output_string.set("Please enter 20 or fewer integers.")
        elif string_counter(n_list) == False:
            self.output_string.set("Please enter positive integers less than 1 quintillion.")
        else:
            if __name__ == "__main__":
                pool = multiprocess.Pool()
                inputs = n_list
                outputs = pool.map(collatz, inputs)
                pool.close()
                pool.join()
            i = 0
            output_string1 = ""
            while i < len(outputs):
                if i == 0:
                    output_string1 = output_string1 + "The hailstone sequence for "
                    output_string1 = output_string1 + str(outputs[i][0])
                    output_string1 = output_string1 + " has a total stopping time of "
                    output_string1 = output_string1 + str(len(outputs[i]) - 1)
                    output_string1 = output_string1 + " and an upper bound of "
                    output_string1 = output_string1 + str(max(outputs[i]))
                    output_string1 = output_string1 + "."
                else:
                    output_string1 = output_string1 + "\nThe hailstone sequence for "
                    output_string1 = output_string1 + str(outputs[i][0])
                    output_string1 = output_string1 + " has a total stopping time of "
                    output_string1 = output_string1 + str(len(outputs[i]) - 1)
                    output_string1 = output_string1 + " and an upper bound of "
                    output_string1 = output_string1 + str(max(outputs[i]))
                    output_string1 = output_string1 + "."
                i += 1
            self.output_string.set(output_string1)
        
    def help1(self):
        self.window1 = tk.Toplevel()
        self.window1.title("Help")
        self.window1.geometry("1250x500")
        self.help_string = '''To begin input up to twenty integers, separated by commas,
         between 1 and 999999999999999999. Click submit and the total stopping times and
         upper bounds will be printed in the app. The program uses multiprocessing, so the
         runtime only depends on the maximum total stopping time, not on the number of
         integers entered.'''
        self.help_label = tk.Label(self.window1, text = self.help_string, font = "Calibri 14")
        self.help_label.pack()
        self.help_window_button = tk.Button(self.window1, text = "Close",
         command = self.close1)
        self.help_window_button.pack()
        
    def close1(self):
        self.window1.destroy()
        
    def acknowledgement(self):
        self.window2 = tk.Toplevel()
        self.window2.title("Acknowledgements")
        self.window2.geometry("800x500")
        self.acknowledgement_string = '''I would like to thank the creators of the
         multiprocess package. It was valuable in reducing the programs' runtime.

         Furthermore, I thank the owner of the YouTube channel Clear Code. His tkinter
         tutorial helped me learn how to create user interfaces in Python. Including the one
         for this project.'''
        self.acknowledgement_label = tk.Label(self.window2,
         text = self.acknowledgement_string, font = "Calibri 14")
        self.acknowledgement_label.pack()
        self.acknowledgement_window_button = tk.Button(self.window2, text = "Close",
         command = self.close2)
        self.acknowledgement_window_button.pack()
        
    def close2(self):
        self.window2.destroy()
        
App("Collatz", "zoomed")
```
